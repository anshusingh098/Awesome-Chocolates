```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Awesome Chocolates.xlsx - Data.csv') 
```

```python
df.head(15)
```

```python
# df = df.loc[df.first_valid_index():]     ## row's delete 
df = df.dropna(axis=0, how='all')  

# df = df.loc[:, df.notna().any()]       ## column's delete
df = df.dropna(axis=1, how='all')  
```

```python
df
```

```python
header_row = df.first_valid_index()
df.columns = df.loc[header_row]

df
```

```python
df = df.reset_index(drop=True)    ## index reset kiya... 0 se start
df
```

```python
df.columns.name = None        ## header ka index no. delete karta hy
df
```

```python
# df = df.iloc[1:]      #method1--- jitna num(jaise 1 type kiya) likhuga utni row hata dega 
df = df.drop(0)         #method2--- ye 0 index wali row delete kr dega 
df
```

```python
df['Date'] = pd.to_datetime(df['Date'],format='mixed')   ## ye tab kam karega jab ek bhi Nun na ho...
df['Date']

df

# df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  ## isme Nun hone pr bhi format change kr dega...
# df['Date']  
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
df.fillna('no_data', inplace=True) 
```

```python
df.head()
```

```python
df['Amount'] = df['Amount'].str.replace('$', '', regex=False)
df['Amount'] = df['Amount'].str.replace(',', '', regex=False).astype(int)
```

```python
df.info()
```

```python
df['Boxes'] = df['Boxes'].astype(int)
df.info()
```

```python
GeographyAmount = df.groupby('Geography')['Amount'].count()

df.groupby('Geography')['Amount'].count()
```

```python
x = GeographyAmount.index
y = GeographyAmount.values

plt.bar(x,y , color="#14C3C9", width=0.8,)
plt.show

plt.title('Geography_Amount ka sum', fontsize=20, fontweight='bold', color="#6D0505", rotation=0 )
plt.xlabel('Geography', fontweight='bold', color="#02343D")
plt.ylabel('Amount', fontweight='bold', color="#F00808")
```

```python
GeographyAmount = df.groupby('Geography')['Amount'].mean()


x = GeographyAmount.index
y = GeographyAmount.values

plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)      #row  #column #position 
plt.bar(x, y)
plt.title('Bar Chart',fontsize=15,fontweight='bold', color="#E41145")
plt.xlabel('Geography', fontweight='bold', color="#B97418")
plt.ylabel('Amount', fontweight='bold', color="#3F0505")

plt.subplot(1, 2, 2)  
plt.plot(x, y, color="#E99B0A", marker='o', markersize=10, linewidth=2, markerfacecolor="#060707", markeredgecolor='black', markeredgewidth=1)
plt.title('Line Chart',fontsize=15,fontweight='bold', color="#CFC20D")
plt.xlabel('Geography', fontweight='bold', color="#230772")
plt.ylabel('Amount', fontweight='bold', color="#3D0606")

plt
plt.tight_layout()
plt.show()
```

```python
GeographyAmount = df.groupby('Geography')['Amount'].max()


x = GeographyAmount.index
y = GeographyAmount.values

plt.figure(figsize=(10,5))

plt.subplot(1, 3, 1)         #row  #column #position 
plt.bar(x, y)
plt.title('chart 1',fontweight='bold', color="#52054E")
plt.xlabel('Geography', fontweight='bold', color="#02343D")
plt.ylabel('Amount', fontweight='bold', color="#F00808")


plt.subplot(1, 3, 2)
plt.pie(y, labels=x, autopct='%1.1f%%')
plt.title('chart 2',fontweight='bold', color="#24053D")


plt.subplot(1, 3, 3)
plt.plot(x, y, color="#E99B0A", marker='o', markersize=10, linewidth=2, markerfacecolor="#060707", markeredgecolor='black', markeredgewidth=1)
plt.title('chart 3',fontweight='bold', color="#5A2208")
plt.xlabel('Geography', fontweight='bold', color="#02343D")
plt.ylabel('Amount', fontweight='bold', color="#F00808")

plt.tight_layout()
plt.show()
```
